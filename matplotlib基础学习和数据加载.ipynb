# 一、matplotlib 基础学习

####

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
```

```python
x
```

```python
y=np.sin(x)
y
```

```python
plt.plot(x,y)### x对应横坐标，y对应纵坐标
```

```python
cy=np.cos(x)
```

```python
siny=y.copy()
```

```python
plt.plot(x,siny,color='red',linestyle="-") 
###颜色选择红色，线条选择‘:’,"-","--"
plt.plot(x,cy)
plt.xlim(-5,15)   ###控制x轴的范围
plt.ylim(0,1.5)####控制y轴的范围
```

```python
plt.plot(x,siny,color='red',linestyle=':')
plt.plot(x,cy)
plt.xlabel("x axis")###给X轴命名
plt.ylabel("y value")###给y轴命名
```

```python
plt.plot(x,siny,color='red',linestyle=':',label="sin(x)")
plt.plot(x,cy,label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()####plot中加标签给线条命名
```

```python
plt.plot(x,siny,color='red',linestyle=':',label="sin(x)")
plt.plot(x,cy,label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("welcom to ML world!")####给整个图命名
```

# 二、读取数据和简单的数据探索

```python
import matplotlib.pyplot as plt
import numpy as np
from  sklearn import datasets 
```

```python
iris=datasets.load_iris()
```

```python
iris.data
```

```python
iris.feature_names  ####数据集的特征名
```

```python
iris.target####对应的数据集中的分类
```

```python
iris.target.shape
```

```python
iris.target_names
```

```python
x=iris.data[:,:2]
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
y=iris.target
```

```python
plt.scatter(x[y==0,0],x[y==0,1],color="red")
####  取 数据集中target等于0的第一列和第二列
plt.scatter(x[y==1,0],x[y==1,1],color="blue")
plt.scatter(x[y==2,0],x[y==2,1],color="green")
```

```python
plt.scatter(x[y==0,0],x[y==0,1],color="red",marker="o") ###设置点的形状
####  取 数据集中target等于0的第一列和第二列
plt.scatter(x[y==1,0],x[y==1,1],color="blue",marker="+")
plt.scatter(x[y==2,0],x[y==2,1],color="green",marker="x")
```

```python
x=iris.data[:,2:]
x.shape
```

```python
plt.scatter(x[y==0,0],x[y==0,1],color="red",marker="o") ###设置点的形状
####  取 数据集中target等于0的第一列和第二列
plt.scatter(x[y==1,0],x[y==1,1],color="blue",marker="+")
plt.scatter(x[y==2,0],x[y==2,1],color="green",marker="x")
```
