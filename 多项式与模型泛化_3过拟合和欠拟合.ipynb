# 过拟合he欠拟合

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
np.random.seed(666)
x = np.random.uniform(-3.0,3.0,size = 100)
X = x.reshape(-1, 1)
y = 0.5 * x**2 + x + 2 + np.random.normal(0,1,size = 100)
```

```python
plt.scatter(x,y)
plt.show()
```

# 使用线性回归

```python
from sklearn.linear_model import LinearRegression  #### 使用线性回归来试试

lin_reg = LinearRegression()
lin_reg.fit(X,y)
lin_reg.score(X,y)
```

```python
y_predict = lin_reg.predict(X)
plt.scatter(x, y)
plt.plot(np.sort(x), y_predict[np.argsort(x)], color='r')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error ##  使用军方误差
```

```python
y_predict = lin_reg.predict(X)
mean_squared_error(y,y_predict)
```

#  多项式回归

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures 
```

```python
def PolynomialRegression(degree): ##对管道进行了一个包装
    return Pipeline([
    ("poly", PolynomialFeatures(degree=degree)),
    ("std_scaler", StandardScaler()),
    ("lin_reg", LinearRegression())
                    ])
```

```python
ploy2_reg = PolynomialRegression(degree=2)
ploy2_reg.fit(X,y)
```

```python
y2_predict = ploy2_reg.predict(X)
mean_squared_error(y,y2_predict)
```

```python
plt.scatter(x, y)
plt.plot(np.sort(x), y2_predict[np.argsort(x)], color='r')
plt.show()
```

```python
ploy10_reg = PolynomialRegression(degree=10)
ploy10_reg.fit(X,y)

y10_predict = ploy10_reg.predict(X)
mean_squared_error(y,y10_predict)
```

```python
plt.scatter(x, y)
plt.plot(np.sort(x), y10_predict[np.argsort(x)], color='r')
plt.show()
```

```python
ploy100_reg = PolynomialRegression(degree=100)
ploy100_reg.fit(X,y)

y100_predict = ploy100_reg.predict(X)
mean_squared_error(y,y100_predict)
```

```python
plt.scatter(x, y)
plt.plot(np.sort(x), y100_predict[np.argsort(x)], color='r')
plt.show()
```

```python
x_plot = np.linspace(-3,3,100).reshape(100,1)
y_plot = ploy100_reg.predict(x_plot)

plt.scatter(x,y)
plt.plot(x_plot[:,0],y_plot,color='r')
plt.axis([-3,3,-1,10])
plt.show()
```
